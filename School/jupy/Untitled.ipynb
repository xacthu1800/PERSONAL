```python
#cau 3
import matplotlib.pyplot as plt
from sklearn import datasets
iris = datasets.load_iris()

# Extract sepal length and sepal width
x = iris.data[:,0]
y = iris.data[:,1]
species = iris.target
# Calculate plot limits
x_min, x_max=x.min()-.5,x.max()+.5
y_min, y_max=y.min()-.5,y.max()+.5

plt.figure()
plt.title('iris classification')
plt.scatter(x,y,c=species)
plt.xlim(x_min,x_max)
plt.ylim(y_min,y_max)
plt.xlabel('length of sepal')
plt.ylabel('width of sepal')
plt.xticks(())
plt.yticks(())
plt.show()
```

```python
#cau 4
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D

%matplotlib notebook

iris = datasets.load_iris()

#x_reduced = PCA(n_components=3).fit_transform(iris.data)
x_reduced = PCA(n_components=3).fit_transform(iris.data[:, [0, 1, 3]])

species = iris.target

fig = plt.figure() 
ax = Axes3D(fig) 
ax.set_title('3D PCA of iris classification')
ax.scatter(x_reduced[:,0], x_reduced[:,1], x_reduced[:,2], c=species)
ax.set_xlabel('component 1')
ax.set_ylabel('component 2')
ax.set_zlabel('component 3')
plt.show()
```

```python
import sklearn
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits

digits = load_digits()
x_digits,y_digits=digits.data, digits.target
n_row,n_col=2,5

def print_digits(image,y):
    fig = plt.figure(figsize=(2*n_col,2*n_row))
    i = 0
    while i < image.shape[0]:
        p = fig.add_subplot(n_row,n_col,i+1,xticks=[],yticks=[])
        p.imshow(image[i],cmap=plt.cm.bone,interpolation='nearest')
        p.text(0,-1,str(y[i]))
        i=i+1
        
print_digits(digits.image, digits.target)
plt.show()
```

```python
# UNSUPERVISED LEARNING cau 2 

import matplotlib.pyplot as plt
from sklearn.datasets import load_digits

digits = load_digits()
x_digits, y_digits = digits.data, digits.target
n_row, n_col = 2, 5

def print_digits(images, y, max_n=10):
    fig = plt.figure(figsize=(2 * n_col, 2 * n_row))
    i = 0
    while i < max_n and images.shape[0]:
        p = fig.add_subplot(n_row, n_col, i + 1, xticks=[], yticks=[])
        p.imshow(images[i].reshape((8, 8)), cmap=plt.cm.bone, interpolation='nearest')
        p.text(0, -1, str(y[i]))
        i = i + 1

print_digits(digits.images, digits.target)
plt.show()
```
